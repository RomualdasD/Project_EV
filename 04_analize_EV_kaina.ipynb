## Autoplius.lt duomenų analizė

```python
# importuojame bibliotekas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Nuskaitome duomenis iš failo
df = pd.read_csv('autoplius_EV_1.csv')
df
```

```python
# Konvertuojame str į int
df['Kaina'] = pd.to_numeric(df['Kaina'], errors='coerce')
```

```python
# Grupuojame pagal modelį
df1 = df.groupby('Modelis').count()['Metai'].sort_values(ascending=False)[:20]
df1
```

```python
# Grafikas. Elektrinių automobilių kiekis pagal modelius, Autoplius.lt
df1.plot(kind='bar', figsize=(16, 9))
plt.xlabel('MODELIS')
plt.ylabel('KIEKIS')
plt.title('Elektrinių automobilių kiekis pagal modelius, Autoplius.lt')
plt.show()
```

```python
# Grupuojame pagal metus
df1 = df.groupby('Metai').count()['Modelis']
df1
```

```python
# Elektrinių automobilių kiekis pagal gamybos metus, Autoplius.lt
df1.plot(kind='bar', figsize=(16, 9))
plt.xlabel('Modelis')
plt.ylabel('Kiekis')
plt.title('Elektrinių automobilių kiekis pagal gamybos metus, Autoplius.lt')
plt.show()
```

```python
# Grupuojame pagal modelį
df1 = df.groupby('Modelis')['Kaina'].mean().sort_values(ascending=False)[:10].round(0)
df1
```

```python
# Brangiausi Elektriniai automobiliai, Autoplius.lt
df1.plot(kind='bar', figsize=(16, 9))
plt.xlabel('MODELIS')
plt.ylabel('KAINA')
plt.title('Brangiausi Elektriniai automobiliai, Autoplius.lt')
plt.show()
```

```python
# Pigiausi Elektriniai automobiliai, Autoplius.lt
df1 = df.groupby('Modelis')['Kaina'].mean().sort_values(ascending=True)[:10].round(0)
df1
```

```python
# Pigiausi Elektriniai automobiliai, Autoplius.lt
df1.plot(kind='bar', figsize=(16, 9))
plt.xlabel('MODELIS')
plt.ylabel('KAINA')
plt.title('Pigiausi Elektriniai automobiliai, Autoplius.lt')
plt.show()
```
